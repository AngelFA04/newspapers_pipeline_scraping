## Describing with graphs

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
sample_elpais = clean_animal.sample(40).reset_index()
sample_animal = clean_animal.sample(40).reset_index()
```

```python
plt.ylabel('Number of tokens in the title')
sample_animal['n_tokens_title'].sample(40).plot(style='k.')
sample_elpais['n_tokens_title'].sample(40).plot(style='r.')
#plt.show()
```

```python
sample_animal['n_tokens_body'].sample(40).plot(style='k.')
sample_elpais['n_tokens_body'].sample(40).plot(style='r.')
```
